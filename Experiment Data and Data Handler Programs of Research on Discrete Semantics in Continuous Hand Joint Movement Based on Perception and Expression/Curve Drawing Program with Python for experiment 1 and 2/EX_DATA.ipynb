```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def Line(x,a,b):
    return x*a+b
def POIA(a1,a2,b1,b2):
    return -((b1-b2)/(a1-a2))
```

```python
FEX_0_a=0.04971
FEX_0_b=-1.483
FEX_1_a=-0.06574
FEX_1_b=1.213
FEX_x=np.arange(0,60,0.01)
FEX_y1=Line(FEX_x,FEX_0_a,FEX_0_b)
FEX_y2=Line(FEX_x,FEX_1_a,FEX_1_b)
FEX_Angle=POIA(FEX_0_a,FEX_1_a,FEX_0_b,FEX_1_b)
```

```python
l1,=plt.plot(FEX_x,FEX_y1,c="#014D64")
l2,=plt.plot(FEX_x,FEX_y2,c="#01A2D9",linestyle='dashed')
plt.axhline(y=0.67, c="#014D64", ls=":", lw=0.75)
plt.axvline(x=FEX_Angle, c="#014D64", ls=":", lw=0.75)
plt.axvspan(xmin=0,xmax=FEX_Angle,facecolor='#014D64',alpha=0.05)
plt.axvspan(xmin=FEX_Angle,xmax=60,facecolor='#ADADAD',alpha=0.05)
plt.yticks([-2,-1,0,0.67,1,2])
plt.annotate("Demarcation Angle\n"+("%.1f" %FEX_Angle)+"$^\circ$",(FEX_Angle+1,-2.5),(FEX_Angle+1,-2.5),size=18)
plt.xlabel("Movement Angle ($^\circ$)",size=18)
plt.ylabel("Average Z-score",size=18)
#plt.title("Finger Abduction",size=18)
plt.legend(handles=[l1,l2],labels=['Standard gesture Movement Angle : 0$^\circ$','Standard gesture Movement Angle : 60$^\circ$'],loc='upper center',bbox_to_anchor=(0.495,-0.22),fontsize=18) #best
plt.tick_params(labelsize=18)
plt.savefig('FEX.png', dpi=600, bbox_inches = 'tight',figsize=(4, 3))
plt.show
```
